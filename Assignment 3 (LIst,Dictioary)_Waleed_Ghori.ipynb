```python
#1. Make a calculator using addition, subtraction,
#multiplication,division and power.

_num1 = int(input("enter Your First Number\n"))
_num2 = int(input("Enter Your Second Number\n"))
_operands = ["+","-","*","/","%","**"]
#print(_operaands)

for opr in _operands:
    if(opr ==  _operands[0]):
        print("Addition: " , _num1 + _num2)
    if(opr ==  _operands[1]):
        print("Subtraction: " , _num1 - _num2)
    if(opr ==  _operands[2]):
        print("Multiplication: " , _num1 * _num2)
    if(opr ==  _operands[3]):
        print("Division: " , _num1 / _num2)
    if(opr ==  _operands[4]):
        print("Modulus: " , _num1 % _num2)
    if(opr ==  _operands[5]):
        print("Power: " , _num1 ** _num2)

```

```python
#5. Write a program to identify duplicate values from list.
import collections
list = [1,2,3,1,5,2,"Ali","Saad",1,"Ali",3,8]
print([item for item, count in collections.Counter(list).items() if count > 1])

```
