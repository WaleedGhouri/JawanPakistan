```python
# Write a python program which takes 5 inputs from user for different subject's marks, total and generate mark sheet using grade ?

Eng = int(input("Enter English Marks:"))
Isl = int(input("Enter Isl Marks:"))
Math = int(input("Enter Maths Marks:"))
Urdu = int(input("Enter Urdu Marks:"))
Comp = int(input("Enter Comp Marks:"))
obtain = Eng + Isl + Math + Urdu + Comp
print("Your Obtain Marks is : ",obtain)
perc = obtain / 500 * 100
print("Your Percentage is : ", perc)
if perc >=80 and perc <=100:
    print("A-One Grade")
elif perc >= 70 and perc < 80:
    print("A Grade")
elif perc >= 60 and perc < 70:
    print("B Grade")
elif perc >= 50 and perc < 60:
    print("C Grade")
elif perc >= 40 and perc < 50:
    print("D Grade")
elif perc >= 33 and perc < 40:
    print("E Grade")
else:
    print("Fail")
```

```python
# Make a Calculator

#Calculator App
num1 = int(input("Enter your First Number"))
num2 = int(input("Enter Your Second Number"))
sum = num1 + num2
sub = num1 - num2
mul = num1 * num2
div = num1 / num2
mod = num1 % num2
 
print("Sum:", sum)
print("Sub:", sub)
print("Mul:", mul)
print("Div:", div)
print("mod:", mod)

#OR

value1 = int(input("Enter your First Value"))
value2 = int(input("Enter your Second Value"))
operand = input("Enter Your Operand") 

if operand == "+":
     print("Sum: ", value1 + value2)
elif operand == "-":
    print("Sub: ", value1 - value2)
elif operand == "*":
    print("Mul: ", value1 * value2)
elif operand == "/":
    print("Div: ", value1 / value2)
elif operand == "%":
    print("Mod: ", value1 % value2)
else:
    print(" Invalid Value ")



```
